## Anomaly Detection: Overview

**Anomaly Detection**, also known as outlier detection, is the process of identifying rare items, events, or observations that differ significantly from the majority of the data. These anomalies can indicate critical incidents, errors, or potential opportunities and are typically rare compared to the normal instances.

### Applications of Anomaly Detection

1. **Fraud Detection**:
   - **Credit Card Fraud**: Detecting unusual spending patterns that may indicate fraudulent transactions.
   - **Insurance**: Identifying suspicious claims that deviate from standard patterns.

2. **Network Security**:
   - **Intrusion Detection**: Identifying unusual network traffic patterns that may indicate cyber-attacks or unauthorized access.
   - **Malware Detection**: Spotting anomalous behavior in software systems that could signify malware infections.

3. **Industrial and Manufacturing**:
   - **Equipment Monitoring**: Detecting unusual behavior in machinery that could predict failures or maintenance needs.
   - **Quality Control**: Identifying defective products in manufacturing processes.

4. **Healthcare**:
   - **Patient Monitoring**: Detecting abnormal vital signs that may indicate a medical emergency.
   - **Medical Imaging**: Identifying unusual patterns in medical images that could signify disease.

5. **Finance and Banking**:
   - **Market Analysis**: Spotting unusual trading activities that could signal market manipulation or insider trading.
   - **Risk Management**: Identifying outliers in financial transactions to assess potential risks.

6. **Environment and Public Safety**:
   - **Environmental Monitoring**: Detecting unusual environmental patterns that may indicate natural disasters or climate changes.
   - **Public Safety**: Identifying unusual behavior in public spaces that could indicate security threats.

### Benefits of Anomaly Detection

1. **Early Detection**:
   - **Preventive Measures**: Early identification of anomalies can help prevent potential disasters or fraud by allowing timely intervention.
   - **Maintenance**: In industrial settings, early detection of equipment anomalies can reduce downtime and maintenance costs.

2. **Improved Security**:
   - **Cybersecurity**: Anomaly detection can enhance security by identifying and mitigating threats before they cause significant harm.
   - **Fraud Prevention**: Detecting and stopping fraudulent activities early can save significant financial losses.

3. **Enhanced Quality Control**:
   - **Product Quality**: In manufacturing, anomaly detection ensures that defective products are identified and removed from the production line.
   - **Service Quality**: In service industries, detecting anomalies in customer behavior can help improve service quality and customer satisfaction.

4. **Resource Optimization**:
   - **Efficient Monitoring**: Automated anomaly detection systems can monitor large volumes of data continuously, reducing the need for manual inspection and freeing up resources for other tasks.
   - **Cost Savings**: Identifying and addressing anomalies early can save costs associated with fraud, maintenance, and errors.

5. **Insight Discovery**:
   - **Data Insights**: Anomaly detection can reveal hidden patterns and insights in data that may not be apparent through standard analysis.
   - **Opportunity Identification**: Identifying unusual patterns can also highlight potential opportunities for business improvement or innovation.

### Techniques for Anomaly Detection

1. **Statistical Methods**:
   - **Z-Score**: Measures the number of standard deviations a data point is from the mean.
   - **Boxplot Method**: Uses the interquartile range (IQR) to identify outliers.

2. **Machine Learning Algorithms**:
   - **Supervised Learning**: Requires labeled data to train models, e.g., classification algorithms.
   - **Unsupervised Learning**: Does not require labeled data, e.g., clustering algorithms like K-Means, DBSCAN.
   - **Semi-Supervised Learning**: Uses a small amount of labeled data along with a large amount of unlabeled data.

3. **Distance-Based Methods**:
   - **K-Nearest Neighbors (KNN)**: Identifies anomalies based on the distance to the nearest neighbors.
   - **Isolation Forest**: An ensemble method that isolates anomalies by randomly selecting a feature and splitting the data.

4. **Density-Based Methods**:
   - **Local Outlier Factor (LOF)**: Identifies anomalies based on the local density deviation of a data point compared to its neighbors.
   - **Density-Based Spatial Clustering of Applications with Noise (DBSCAN)**: Detects anomalies by identifying low-density regions.

### Challenges in Anomaly Detection

1. **Imbalanced Data**:
   - Anomalies are often rare compared to normal instances, making it challenging to train models effectively.

2. **High Dimensionality**:
   - In high-dimensional spaces, detecting anomalies becomes difficult due to the curse of dimensionality.

3. **Dynamic Data**:
   - Anomalies can change over time, requiring models to adapt to new patterns and trends.

4. **Noise in Data**:
   - Distinguishing between true anomalies and noise can be challenging, especially in noisy datasets.

5. **Scalability**:
   - Efficiently processing large datasets to detect anomalies requires scalable algorithms and infrastructure.

### Conclusion

Anomaly detection is a critical component in many applications across various industries, providing significant benefits in terms of early detection, improved security, enhanced quality control, resource optimization, and insight discovery. ## Categories of Anomaly Detection Methods

1. **Statistical Methods**:
   - **Z-Score**: Measures the number of standard deviations a data point is from the mean.
   - **Boxplot Method (IQR)**: Uses the interquartile range to identify outliers.

2. **Machine Learning-Based Methods**:
   - **Isolation Forest**: An ensemble method that isolates anomalies by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature.
   - **Local Outlier Factor (LOF)**: Identifies anomalies by comparing the local density of a data point to that of its neighbors.
   - **Elliptic Envelope**: Fits a Gaussian distribution to the data and identifies anomalies based on the Mahalanobis distance.

3. **Distance-Based Methods**:
   - **K-Nearest Neighbors (KNN)**: Identifies anomalies based on the distance to the nearest neighbors. Data points that are far away from their nearest neighbors are considered anomalies.

4. **Density-Based Methods**:
   - **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Identifies anomalies as points that lie in low-density regions.
   - **OPTICS (Ordering Points To Identify the Clustering Structure)**: Extends DBSCAN to handle varying densities. Identifies anomalies based on the reachability distance.

5. **Ensemble Methods**:
   - **Isolation Forest**: Also falls into this category, as it is an ensemble method that builds multiple decision trees to isolate anomalies.
   - **Random Forest for Outlier Detection**: Uses an ensemble of decision trees to identify anomalies.

### Comparison of Methods

1. **Statistical Methods**:

### 1: Boxplot Method (IQR)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml

# Load the Adult Census Income dataset from OpenML
adult = fetch_openml(data_id=1590, as_frame=True, parser='auto')
df = adult.frame

# Rename columns to lower case and replace hyphens with underscores
df.columns = [col.lower().replace('-', '_') for col in df.columns]

# Select numeric features
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
X_numeric = df[numeric_features]

# Function to detect outliers using IQR
def detect_outliers_iqr(data, features):
    outliers = pd.Series([False] * len(data))
    for feature in features:
        Q1 = data[feature].quantile(0.25)
        Q3 = data[feature].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        feature_outliers = (data[feature] < lower_bound) | (data[feature] > upper_bound)
        outliers = outliers | feature_outliers
    return outliers

# Detect outliers using IQR
outliers_iqr = detect_outliers_iqr(X_numeric, numeric_features)

# Add outlier labels to the original DataFrame
df['outlier_iqr'] = outliers_iqr

# Plot boxplots for numeric features with outliers
plt.figure(figsize=(10,6))
sns.boxplot(data=X_numeric, palette="PuRd")
plt.xticks(rotation=90)
plt.title('Boxplot of Numeric Features with Outliers Detected by IQR Method')
plt.show()
```

### 2: Z-Score Method

1. **Z-Score Method**:
   - **Standardization**: Standardize the numeric features to have zero mean and unit variance.
   - **Calculate Z-Scores**: Compute the absolute Z-scores for each data point.
   - **Identify Outliers**: Data points with Z-scores greater than a threshold (e.g., 3) are considered outliers.
   - **Visualization**: Plot the detected outliers.

2. **Boxplot Method (IQR)**:
   - **Calculate IQR**: Compute the interquartile range (IQR) for each numeric feature.
   - **Identify Outliers**: Data points outside the range \([Q1 - 1.5 \times IQR, Q3 + 1.5 \times IQR]\) are considered outliers.
   - **Visualization**: Plot the detected outliers.

### Conclusion

These two statistical methods provide a baseline for outlier detection. By comparing their results with more sophisticated anomaly detection techniques, you can evaluate the effectiveness of each method and determine the best approach for your data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_openml

# Load the Adult Census Income dataset from OpenML
adult = fetch_openml(data_id=1590, as_frame=True, parser='auto')
df = adult.frame

# Rename columns to lower case and replace hyphens with underscores
df.columns = [col.lower().replace('-', '_') for col in df.columns]

# Select numeric features
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
X_numeric = df[numeric_features]

# Standardize the numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

# Calculate Z-scores
z_scores = np.abs(X_scaled)

# Define a threshold for Z-scores
threshold = 3

# Identify outliers
outliers = (z_scores > threshold).any(axis=1)

# Add outlier labels to the original DataFrame
df['outlier_zscore'] = outliers

def plot_anomalies(df, anomaly_col, colors, method_name):
    # Determine the xticks based on unique values in anomaly_col
    unique_values = sorted(df[anomaly_col].unique())
    xticks = range(len(unique_values))

    # Mapping for x-ticks labels
    xticks_labels = ['Normal' if val == False else 'Outlier' for val in unique_values]

    # Plot the anomalies
    plt.figure(figsize=(9, 5))
    sns.scatterplot(x=df.index, y=[0 if label else 1 for label in df[anomaly_col]], hue=df[anomaly_col], palette=colors, marker='+')
    plt.legend(title='Anomaly', loc='upper left')
    plt.xlabel('Index')
    plt.ylabel('Anomaly Label')
    plt.title(f'{method_name} Anomaly Detection')
    plt.yticks([0, 1], ['Outlier', 'Normal'])
    plt.show()

    # Create a countplot of the anomaly column
    plt.figure(figsize=(9, 5))
    sns.countplot(x=anomaly_col, data=df, hue=anomaly_col, palette=colors)
    plt.title(f'Count of Outliers vs Normal Points ({method_name})')
    plt.xlabel('Anomaly Label')
    plt.ylabel('Count')
    plt.xticks(ticks=xticks, labels=xticks_labels)
    plt.legend(title='Anomaly')
    plt.show()

    # Display the first few rows of the DataFrame with anomaly labels
    cols = ['age', 'education', 'occupation', 'race', 'sex', 'class'] + [anomaly_col]
    print(df[cols].head())

    # Print the count of anomalies and normal points
    print(df[anomaly_col].value_counts())

# Define the colors
colors = ['purple', 'red']

# Call the function to plot anomalies for Z-Score method
plot_anomalies(df, 'outlier_zscore', colors, 'Z-Score Method')
```

### 3: Isolation Forest

Isolation Forest is an ensemble method that isolates anomalies by randomly selecting a feature and then randomly selecting a split value between the maximum and minimum values of the selected feature.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import IsolationForest
from sklearn.datasets import fetch_openml

# Load the Adult Census Income dataset from OpenML
adult = fetch_openml(data_id=1590, as_frame=True, parser='auto')
df = adult.frame

# Rename columns to lower case and replace hyphens with underscores
df.columns = [col.lower().replace('-', '_') for col in df.columns]

# Select features and target
X = df.drop(columns=['class'])

# Take a sample of the data to avoid running out of memory
sample_size = 0.5
X_sample = X.sample(frac=sample_size, random_state=42)
df_sample = X_sample.copy()

# Identify column types
numeric_features = X_sample.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_sample.select_dtypes(include=['object', 'category']).columns.tolist()

# Define preprocessing for numeric columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

# Define preprocessing for categorical columns
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])

# Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Create the pipeline with Isolation Forest
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('isolation_forest', IsolationForest(contamination=0.1, random_state=42))
])

# Fit the pipeline
pipeline.fit(X_sample)

# Preprocess the data
X_sample_preprocessed = pipeline.named_steps['preprocessor'].transform(X_sample)

# Predict anomalies
anomaly_labels = pipeline.named_steps['isolation_forest'].predict(X_sample_preprocessed)
# Anomalies are labeled as -1, normal points are labeled as 1
df_sample['anomaly'] = anomaly_labels

# Function to plot anomalies
def plot_anomalies(df, anomaly_col, colors, method_name):
    unique_values = sorted(df[anomaly_col].unique())
    xticks = range(len(unique_values))
    xticks_labels = ['Normal' if val == 1 else 'Outlier' for val in unique_values]

    plt.figure(figsize=(9, 4))
    sns.scatterplot(x=df.index, y=df[anomaly_col], hue=df[anomaly_col], palette=colors, marker='+')
    plt.legend(title='Anomaly', loc='upper left')
    plt.xlabel('Index')
    plt.ylabel('Anomaly Label')
    plt.title(f'{method_name} Anomaly Detection')
    plt.show()

    plt.figure(figsize=(9, 4))
    sns.countplot(x=anomaly_col, data=df, hue=anomaly_col, palette=colors, alpha=0.8)
    plt.title(f'Count of Outliers vs Normal Points ({method_name})')
    plt.xlabel('Anomaly Label')
    plt.ylabel('Count')
    plt.xticks(ticks=xticks, labels=xticks_labels)
    plt.legend(title='Anomaly')
    plt.show()

    cols = ['age', 'education', 'occupation', 'relationship', 'race', 'sex'] + [anomaly_col]
    print(df[cols].head())
    print('\n', df[anomaly_col].value_counts())

# Define the colors
colors = ['red', 'purple']

# Call the function to plot anomalies for Isolation Forest method
plot_anomalies(df_sample, 'anomaly', colors, 'Isolation Forest')
```
